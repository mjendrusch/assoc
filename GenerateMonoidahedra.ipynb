```python
from matplotlib import pyplot as plt

from assoc.definitions import tree, leaf, unit, comp, rho, lamb
from assoc.plots import sprout_plot, tree_plot
```

```python
# define the expression (\rho_a \otimes id_b) \otimes (id_c \otimes \lambda_d)
# and plot its domain and codomain:

expr = comp(comp(rho, leaf), comp(leaf, lamb))
boundary = expr.boundary()

print(expr)
```

```python
# This gives us the expected domain and codomain,
# mapping $((a \otimes 1) \otimes b) \otimes (c \otimes (1 \otimes d)$
# to $(a \otimes b) \otimes (c \otimes d)$

fig, ax = plt.subplots(1, 2, figsize=(5, 5))
for idx in range(2):
    tree_plot(boundary[idx], ax=ax[idx])
```
